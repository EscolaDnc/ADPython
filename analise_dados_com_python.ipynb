# Pacote Pandas

## **1. O que é o Pandas?**

O Pandas é uma das bibliotecas mais importantes para análise e manipulação de dados em Python. Seu nome vem da expressão *Panel Data* (dados em painéis), um conceito estatístico usado para organizar e processar grandes volumes de dados.

O Pandas é amplamente utilizado em ciência de dados, aprendizado de máquina, estatística e até mesmo em negócios, pois fornece ferramentas eficientes para trabalhar com tabelas, manipular informações e analisar grandes conjuntos de dados de maneira intuitiva.

## **2. Importância do Pandas**

A principal vantagem do Pandas é que ele permite **trabalhar com dados estruturados de forma eficiente**. Antes de seu surgimento, manipular tabelas de dados em Python era complicado, exigindo o uso de listas e dicionários, que não eram otimizados para grandes volumes de informações.

Com o Pandas, é possível:

✔ Carregar e salvar dados de diversos formatos (CSV, Excel, SQL, JSON, etc.).

✔ Filtrar, modificar e transformar tabelas de maneira eficiente.

✔ Realizar cálculos estatísticos e sumarizações de dados.

✔ Trabalhar com grandes volumes de informações sem comprometer o desempenho.

✔ Facilitar a integração com outras bibliotecas, como NumPy e Matplotlib.

Essas características fazem do Pandas uma ferramenta essencial para quem trabalha com dados.

## Series Pandas

Uma **`Series`** é uma das estruturas de dados fundamentais do Pandas. Ela pode ser entendida como uma **coluna de uma tabela**, semelhante a uma lista ou um array do NumPy, mas com funcionalidades adicionais.

### **1. Estrutura de uma `Series`**

Uma `Series` no Pandas é composta por **dois elementos principais**:

1. **Valores (dados)** – O conjunto de informações armazenadas (números, textos, datas, etc.).
2. **Índice** – A identificação de cada elemento dentro da `Series` (por padrão, é numérico e começa do zero, mas pode ser personalizado).

| Índice | Valor (Dados) |

Aqui, os **valores** representam os nomes das frutas, enquanto os **índices** identificam cada item na `Series`.

## DataFrame - Criação e Fatiamento

O **`DataFrame`** é a estrutura de dados mais importante do Pandas. Ele pode ser entendido como uma **tabela**, semelhante a uma planilha do Excel ou uma tabela de banco de dados.

Enquanto uma **`Series`** representa uma **única coluna de dados**, um **`DataFrame`** pode conter **múltiplas colunas**, permitindo armazenar e manipular dados de forma mais completa.

---

### **Estrutura de um `DataFrame`**

Um `DataFrame` é composto por **três elementos principais**:

✔ **Linhas (Índices)** – Identificam cada registro (por padrão, são números sequenciais, mas podem ser personalizados).

✔ **Colunas** – Representam diferentes categorias de dados (como nome, idade, salário, etc.).

✔ **Valores (Dados)** – O conteúdo armazenado dentro do `DataFrame`.

### **Exemplo de um `DataFrame`**

| Índice | Nome | Idade | Cidade | Salário |
| --- | --- | --- | --- | --- |
| 0 | Ana | 25 | São Paulo | 3500 |
| 1 | Bruno | 30 | Rio de Janeiro | 4200 |
| 2 | Carla | 28 | Belo Horizonte | 3800 |

- **Cada linha** representa um **registro** (uma pessoa).
- **Cada coluna** representa um **atributo** (Nome, Idade, Cidade, Salário).
- Os **índices** identificam cada linha, começando de **0** por padrão.

## Seleção Condicional

### **Como Funciona a Seleção Condicional?**

A seleção condicional funciona com **comparações lógicas** que verificam se os valores em uma coluna atendem a um critério específico. Essas comparações geram um "filtro", que retorna apenas as linhas que satisfazem a condição.

### **Principais Comparações Lógicas Usadas na Seleção Condicional**

✔ **Igualdade (`==`)** – Seleciona os registros que possuem um valor exato em determinada coluna.

✔ **Diferença (`!=`)** – Retorna os registros que *não* possuem um valor específico.

✔ **Maior (`>`) e Menor (`<`)** – Filtra registros acima ou abaixo de um determinado valor.

✔ **Maior ou Igual (`>=`) e Menor ou Igual (`<=`)** – Similar às anteriores, mas incluindo o valor de referência.

### **Operadores Lógicos Combinados**

Além das comparações individuais, podemos combinar condições para criar filtros mais complexos:

✔ **E (`&`)** – Exige que *todas* as condições sejam verdadeiras.

✔ **Ou (`|`)** – Retorna registros que atendem *pelo menos uma* das condições.

✔ **Negação (`~`)** – Exclui registros que atendem a uma determinada condição.

## Dados Ausentes

### **1. O Que São Dados Ausentes?**

Quando trabalhamos com **DataFrames do Pandas**, muitas vezes lidamos com conjuntos de dados **incompletos**, ou seja, há valores **faltando** em algumas colunas ou linhas. Esses valores ausentes podem ser representados de diferentes formas, sendo a mais comum o `NaN` (**Not a Number**), que indica que uma determinada célula não possui informação válida.

---

### **2. Por Que os Dados Ausentes São um Problema?**

A presença de valores ausentes pode afetar significativamente a análise de dados. Alguns problemas comuns incluem:

✔ **Erros em Cálculos Estatísticos** – Muitas funções matemáticas e estatísticas podem não funcionar corretamente se houver dados faltando.

✔ **Impacto na Modelagem de Machine Learning** – Algoritmos de aprendizado de máquina podem não aceitar valores ausentes, exigindo tratamento prévio.

✔ **Distorção em Gráficos e Relatórios** – Relatórios e visualizações podem ficar incompletos ou apresentar tendências incorretas.

✔ **Inconsistência nos Dados** – Se algumas colunas têm muitas informações ausentes, pode ser difícil comparar registros ou identificar padrões.

Portanto, é essencial **identificar, analisar e tratar** os dados ausentes antes de prosseguir com qualquer análise.

## Group By

## **1. O Que é o `groupby` no Pandas?**

O método **`groupby`** no Pandas permite agrupar os dados de um `DataFrame` com base em uma ou mais colunas e **realizar cálculos agregados** sobre esses grupos.

Essa função é muito útil quando queremos analisar informações **de forma categorizada**, respondendo perguntas como:

- Qual é o **salário médio** por departamento em uma empresa?
- Quantas **vendas foram realizadas por cidade**?
- Qual é a **nota média dos alunos por turma**?

O `groupby` facilita a análise de dados ao resumir grandes tabelas de maneira organizada.

---

## **2. Exemplo Didático: Entendendo o `groupby`**

Vamos considerar a seguinte tabela de **funcionários de uma empresa**:

| Nome | Departamento | Salário | Cidade |
| --- | --- | --- | --- |
| Ana | TI | 5000 | São Paulo |
| Bruno | TI | 6000 | Rio de Janeiro |
| Carla | RH | 4000 | São Paulo |
| Daniel | RH | 4200 | Belo Horizonte |
| Eduardo | Vendas | 3000 | São Paulo |
| Fernanda | Vendas | 3500 | Rio de Janeiro |

---

## **3. Aplicações do `groupby`**

Agora, vamos ver algumas formas de agrupar os dados dessa tabela para responder perguntas comuns.

### **✔ 1. Qual é o salário médio por departamento?**

Se quisermos calcular o **salário médio** dos funcionários agrupados por **departamento**, teremos:

| Departamento | Salário Médio |
| --- | --- |
| TI | 5500 |
| RH | 4100 |
| Vendas | 3250 |

Aqui, o `groupby` agrupou os funcionários por **departamento** e calculou a média de seus salários.

---

### **✔ 2. Quantos funcionários há por cidade?**

Se quisermos contar quantos funcionários trabalham em cada cidade, teremos:

| Cidade | Número de Funcionários |
| --- | --- |
| São Paulo | 3 |
| Rio de Janeiro | 2 |
| Belo Horizonte | 1 |

Nesse caso, agrupamos os dados por **cidade** e contamos quantos registros existem em cada grupo.

---

### **✔ 3. Qual é o maior salário por departamento?**

Podemos usar o `groupby` para identificar o **maior salário** dentro de cada departamento:

| Departamento | Maior Salário |
| --- | --- |
| TI | 6000 |
| RH | 4200 |
| Vendas | 3500 |

## **Concatenar, Juntar e Mesclar**

### **1. O Que Significa Concatenar e Juntar DataFrames?**

No Pandas, podemos **combinar diferentes DataFrames** para formar uma única estrutura de dados. Para isso, usamos duas principais abordagens:

✅ **Concatenação (`concat`)** – Junta DataFrames **um abaixo do outro** ou **lado a lado**, sem considerar chaves em comum.

✅ **Junção (`merge`)** – Une DataFrames **baseando-se em colunas ou índices comuns**, semelhante a um **JOIN** no SQL.

Essas técnicas são fundamentais quando trabalhamos com conjuntos de dados distribuídos em várias tabelas e precisamos analisá-los juntos.

```python
import numpy as np
import pandas as pd
```

```python
usuarios = {
    "ID_usuario": [1, 2, 3],
    "Nome": ["Maria", "Bruno", "Joao"],
    "Cidade": ["São Paulo", "Rio", "Belo Horizonte"]
}

compras = {
    "ID_usuario": [1, 2, 1, 1, 3, 3, 2, 3],
    "Produto": ["Notebook", "Camera", "Camiseta", "Camiseta", "Caneca", "Caneca", "Calça", "Calça"],
    "Valor": [2000, 1000, 70, 60, 50, 50, 120, 100]
}

usuarios_novos = {
    "ID_usuario": [4, 5],
    "Nome": ["Pedro", "Marcos"]
}
```

```python
usuarios_df = pd.DataFrame(usuarios)
compras_df = pd.DataFrame(compras)
usuarios_novos_df = pd.DataFrame(usuarios_novos)
```

```python
usuarios_df
```

```python
compras_df
```

```python
usuarios_novos_df
```

**Concatenando Dataframes**

```python
pd.concat([usuarios_df, usuarios_novos_df])
```

```python
usuarios_geral = pd.concat([usuarios_df, usuarios_novos_df])
```

**Mesclando DataFrames**

```python
pd.merge(usuarios_geral, compras_df, how="left", on="ID_usuario")
```

```python
pd.merge(usuarios_geral, compras_df, how="right", on="ID_usuario")
```

```python
pd.merge(usuarios_geral, compras_df, how="inner", on="ID_usuario")
```

## Outras Operações

O Pandas é uma biblioteca poderosa para manipulação e análise de dados. Ele fornece diversas operações que facilitam a exploração, organização e transformação dos dados.

Nesta aula, vamos aprender algumas das principais operações usadas em **DataFrames**, que são:

- `head()`
- `unique()`
- `value_counts()`
- `apply()`
- `index`
- `columns`
- `sort_values()`

---

### **1. `head()` – Visualizando as Primeiras Linhas**

O método **`head()`** permite visualizar as **primeiras linhas** de um DataFrame.

🔹 **Para que serve?**

- É útil para inspecionar rapidamente a estrutura e os dados de um DataFrame.
- Ajuda a verificar se os dados foram carregados corretamente.

✔ Por padrão, `head()` exibe **as 5 primeiras linhas**, mas podemos escolher quantas linhas queremos visualizar.

---

### **2. `unique()` – Encontrando Valores Únicos**

O método **`unique()`** retorna **todos os valores distintos** de uma coluna específica.

🔹 **Para que serve?**

- Identificar **categorias** dentro de uma coluna (exemplo: cidades, produtos, cargos).
- Detectar **possíveis valores inválidos ou inconsistentes** nos dados.

✔ Funciona apenas para colunas específicas de um DataFrame.

---

### **3. `value_counts()` – Contando a Frequência dos Valores**

O método **`value_counts()`** conta quantas vezes cada valor aparece em uma coluna.

🔹 **Para que serve?**

- Analisar a distribuição dos dados dentro de uma categoria.
- Saber **quantos clientes são de cada cidade** ou **quantos produtos foram vendidos de cada tipo**.

✔ O resultado vem ordenado do maior para o menor valor.

---

### **4. `apply()` – Aplicando Funções aos Dados**

O método **`apply()`** permite aplicar uma função personalizada **a cada linha ou coluna** de um DataFrame.

🔹 **Para que serve?**

- Aplicar transformações **sem precisar de um loop manual**.
- Criar novas colunas baseadas em cálculos personalizados.
- Alterar valores existentes de forma eficiente.

✔ Pode ser usado com funções **do próprio Pandas ou criadas pelo usuário**.

---

### **5. `index` – Acessando os Índices do DataFrame**

O atributo **`index`** retorna ou modifica o **índice** do DataFrame, que são os números das linhas.

🔹 **Para que serve?**

- Saber quais são os índices atuais do DataFrame.
- Alterar ou redefinir os índices para facilitar a análise.
- Definir uma coluna como índice principal.

✔ O índice pode ser um número (0, 1, 2...) ou um valor específico, como **IDs de clientes ou datas**.

---

### **6. `columns` – Acessando os Nomes das Colunas**

O atributo **`columns`** retorna a lista de nomes das colunas do DataFrame.

🔹 **Para que serve?**

- Identificar rapidamente as colunas disponíveis.
- Renomear colunas ou selecionar apenas algumas delas para análise.

✔ Muito útil para **ajustar nomes de colunas** ao importar dados de diferentes fontes.

---

### **7. `sort_values()` – Ordenando os Dados**

O método **`sort_values()`** permite ordenar o DataFrame com base nos valores de uma ou mais colunas.

🔹 **Para que serve?**

- Ordenar **do maior para o menor** (decrescente) ou **do menor para o maior** (crescente).
- Organizar **dados de vendas** do maior faturamento para o menor.

✔ Podemos ordenar por **uma ou mais colunas ao mesmo tempo**.
